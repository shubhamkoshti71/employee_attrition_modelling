<h3><b>Handling missing values</b></h3>

```python
#Checking for missing values in each dataset
print("Missing Values in Employee Survey Data:")
print(employee_survey_data.isna().sum())
print("\nMissing values in General Data:")
print(general_data.isna().sum())
print("\nMissing Values in Manager Survey Data:")
print(manager_survey_data.isnull().sum())
```

<h4>Handling missing values in employee survey data dataframe</h4>

```python
#Handling missing values in employee_survey_data using mean imputation
cols = employee_survey_data.select_dtypes(np.number).columns
print(cols)

employee_survey_data[cols] = employee_survey_data[cols].fillna(employee_survey_data[cols].mean())
```

```python
#Converting column values to int datatype after rounding to nearest whole number
employee_survey_data_imputed = employee_survey_data.round(0).astype(int)
```

<h4>Handling missing values in general data dataframe</h4>

```python
#In general_data dataset, the column (NumCompaniesWorked) has NaNs and the distribution of this column w.r.t Age is uniform.
plt.bar(general_data['Age'], general_data['NumCompaniesWorked'])

plt.xlabel('Age')
plt.ylabel('Number of companies worked for')

```

```python
#Based on above analysis, it is safe to replace the NaNs in NumCompaniesWorked column with 0

general_data['NumCompaniesWorked'] = general_data['NumCompaniesWorked'].fillna(0).astype(int)


general_data.head()
```

```python
'''Another column in general_data has NaNs (TotalWorkingYears). To replace the NaNs in this column, I used a technique:
I calculated the average number of working years for each value in the NumCompaniesWorked and replaced the NaNs with this corresponding average values'''

# Calculating the average number of years worked (TotalWorkingYears) for each value of NumCompaniesWorked
average_years_worked = general_data.groupby('NumCompaniesWorked')['TotalWorkingYears'].mean().reset_index()

average_years_worked
```

```python
#Here using apply() function, the code replaces the NaNs in TotalWorkingYears with the corresponding average values in the above dataframe(average_years_worked)
general_data['TotalWorkingYears'] = general_data.apply(
    lambda row: average_years_worked[average_years_worked['NumCompaniesWorked'] == row['NumCompaniesWorked']]['TotalWorkingYears'].values[0]
    if pd.isna(row['TotalWorkingYears']) else row['TotalWorkingYears'],
    axis=1
)

#Checking if there are any more NaNs in the general_data dataframe
general_data.isna().sum()
```

```python
#changing datatype of TotalWorkingYears from float to integer
general_data['TotalWorkingYears'] = general_data['TotalWorkingYears'].round(0).astype(int)

#storing the imputed general_data dataframe in a new variable 
general_data_imputed = general_data

general_data_imputed.head(10)
```

<h4>Handling missing values for in_time and out_time data</h4>

```python
#I thought of combining both the in_time and out_time datasets before handling the missing values in both datasets

#Firstly, setting the index to Employee ID for both the datasets
in_time = in_time.rename(columns={'Unnamed: 0': 'EmployeeID'})
out_time = out_time.rename(columns={'Unnamed: 0': 'EmployeeID'})

in_time = in_time.set_index("EmployeeID")
out_time = out_time.set_index("EmployeeID")

#Next, removing the NaN values in both the dataframes by replacing them with 0

in_time = in_time.fillna(0)
out_time = out_time.fillna(0)

#Now converting the columns to datetime except for the EmployeeID column
bad_column = "EmployeeID"
other_cols = out_time.columns.difference([bad_column])
in_time[other_cols] = in_time[other_cols].apply(pd.to_datetime, errors='coerce')
out_time[other_cols] = out_time[other_cols].apply(pd.to_datetime, errors='coerce')

```

```python
#displaying first few elements of in_time dataframe
in_time.head(5)
```

```python
#displaying first few elements of out_time dataframe
out_time.head(5)
```

```python
'''Now to combine both these dataframes, we'll calculate the difference between in and out times for each employee (in hours) and then 
aggregate all the values over the horizontal axis to obtain the total working hours for each employee'''

working_hours = (out_time - in_time).apply(lambda x: x.dt.total_seconds() / 3600)
total_working_hours = working_hours.sum(axis=1).round(0).astype(int)

```

```python
#resetting the index
total_working_hours.reset_index(drop=True, inplace=True)
```

<h3><b> Merging the datasets on a common key </b></h3>

```python
merged_data = general_data_imputed.merge(employee_survey_data_imputed, on= 'EmployeeID')
merged_data = merged_data.merge(manager_survey_data, on='EmployeeID')
merged_data['TotalWorkingHours'] = total_working_hours
```

```python
merged_data.head(10)
```

```python
#converting the dataframe merged_data to a csv file
merged_data.to_csv('cleaned_data.csv', index= False)
```
